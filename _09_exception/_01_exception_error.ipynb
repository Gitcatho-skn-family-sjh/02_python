# 예외-오류 Exception-Error

- Exception : 정상적인 흐름으로 복구가능한 미약한 오류. (try-except구문으로 복구)
- Error : 복구불가한 심각한 오류 (재시작, 소수코드 수정) (OOM Out of Memory)

파이썬의 예외/오류 클래스는 모두 BaseException클래스를 상속.

---
### 조건문과 예외처리 구분
- if-else는 정상적인 선택지 중 하나를 고르는 문법
- exception은 정상 흐름으로 처리하기 어려운 문제가 발생했을 때 흐름을 분리하는 문법

## 예외처리
**(중요) 발생된 예외/오류 상황에서 try..except구문을 통해 적절히 처리후 정상적인 흐름을 돌려내는 것**

```python
print("시작")

#try: 예외가 발생할 가능성이 있는 코드를 시도하는 구문
try:
    result = 10/0 # 예외가 발생할 수 있는 코드

    # 예외 발생 원리
    # 1. 예외 상황 감지
    # 2. 해당 예외 상황에 맞는 "예외 클래스"를 찾음
    # 3. 예외 클래스를 이용해서 "예외 객체"를 생성
    # 4. 생성된 예외 객체를 예외 상황이 감지된 코드 위치로 올림(raise)

except:
    # except: try에서 올려진 예외 처리 구문
    print("0으로 나눌 수 없습니다.")

    # except 구문이 종료되면 다시 정상 흐름으로 복귀
print("종료")
```

```python
# excpet 구문을 이용한 예외 객체 사용 방법
import random # 난수 생성 모듈

try:
    num = random.randint(0,1) # 0~1사이 정수형 난수 생성
    print("num: ", num)
    # 0 == False, 1 == True
    if num: #1인경우
        print('abc' +123)
    else: #0인경우
        print(int('abc'))

# e : try에서 발상한 예외 객체
except TypeError as e:
    print(e)
    print("예외 타입 이름: ",type(e).__name__)
    print("예외 메시지: ",str(e))
    print("예외 상세 값: ",e.args)
    print(f'[Error] {type(e).__name__}: {str(e)}')
except ValueError as e:
    print(e)
    print("예외 타입 이름: ",type(e).__name__)
    print("예외 메시지: ",str(e))
    print("예외 상세 값: ",e.args)
    print(f'[Error] {type(e).__name__}: {str(e)}')
```

```python
# 숫자를 입력 받아
# 입력 받은 숫자와 key가 일치하는 학생을 조회하여 id, name 출력
# 입력 받은 숫자가 students에 없으면 "1~3사이 숫자만 입력해주세요"
# + 잘못된 숫자 입력 시 다시 숫자를 입력 받게 하기

students: dict[int, dict] = {
    1: {"id": 10, "name": "홍길동"},
    2: {"id": 20, "name": "신사임당"},
    3: {"id": 30, "name": "이순신"},
}
while True: # while 조건식 -> 조건식이 True인 경우 반복
    try:
        num: int = int(input("숫자 입력: "))
        student: dict = students[num]
        print('id: ',student['id'])
        print('name: ',student['name'])
        break # 정상입력시 반복 종료

    except KeyError as e:
        print("1~3사이 숫자만 입력해주세요")

print("--- 프로그램 종료 ---")
```

### pass
- if, for, def, class, except 등의 블록은 내부가 비어있으면 오류가 발생함
- 이 때, 구조만 유지하는 용도 또는 예외를 무시할 때 사용

```python
if True:
    pass

def func():
    pass
```

```python
# 예외 무시
try:
    print(10/0)
except:
    pass

print("종료")
```

### finally
- try 구문에서 예외 발생 여부와 관계 없이
  마지막에 무조건 실행하는 구문(블록)


### else
- try 구문에서 예외 미발생 시 수행되는 구문

```python
try:
    num = int(input("숫자 입력"))
    result = 100 / num

except ValueError as e:
    print("정수형 숫자만 입력하세요")
except ZeroDivisionError as e:
    print("0으로 나눌 수 없습니다")

else:
    print("계산 결과 : ", result)
finally:
    print("예외 발생 여부와 관계 없이 무조건 수행")
```

```python
# finally 사용 예시
try :
    f = open('helloworld.txt', 'r', encoding='utf-8')
    print(f.read())

    print(10/0) # 예외 발생 상황

except:
    pass

finally:
    f.close() #메모리 반환(필수)
```

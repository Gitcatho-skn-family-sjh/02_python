```python
## **문제 3: 업무 시간 계산하기**
#
# **"다람쥐는 오늘 근무한 시간을 계산하고 싶다."**
#
# 1. `work_hours(start, end)`라는 함수를 만들어 매개변수 `start`(출근 시간)와 `end`(퇴근 시간)를 입력받아 근무 시간을 계산하세요.
#     - 근무 시간 = `end - start`
# 2. 반환된 근무 시간을 출력하세요.
#     - 출력 예시: **"오늘 근무한 시간은 [근무 시간]시간입니다."**

def work_hours(start, end):
    return end - start

start = int(input("출근 시간을 입력하세요: "))
end = int(input("퇴근 시간을 입력하세요: "))

hours = work_hours(start, end)

print(f"오늘 근무한 시간은 {hours}시간입니다.")
```
